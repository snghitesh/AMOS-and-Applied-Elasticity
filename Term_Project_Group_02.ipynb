# <font color = "green"> Introduction: Levy's Method
According to the Levy's method,
deflection of a plate is given by the following form,
$$ w(x,y) = \sum_{m=1}^{\infty} [Y_m(y)] \sin \frac{{m\pi x}}{{a}} $$
Here, $Y_m(y)$ need to be determined and must satisfy appropriate boundary conditions at $y=0$ and $y=b$.
Substituting above expression of $w(x,y)$ in the total potential energy formula,
    
$$ \pi = \frac{{D}}{{2}}\int\int_{{R}}^{{}} \{(\nabla ^2 w)^2 + 2(1-\nu)[(\frac{{\partial^2 w}}{{\partial x \partial y}})^2 - \frac{{\partial^2 w}}{{\partial x^2}}\frac{{\partial^2 w}}{{\partial y^2}}]\} dxdy - \int\int_{{R}}^{{}} qw \ dxdy $$

We obtain,
$$ \pi = \frac{{D}}{{2}}\int_{{0}}^{{b}}\int_{{0}}^{{a}} \{[\sum_{m=1}^{\infty} (-Y_m(\frac{{m\pi}}{{a}})^2\sin \frac{{m\pi x}}{{a}})
+ Y_m^{{''}}\sin \frac{{m\pi x}}{{a}}]^2 + 2(1-\nu)[\sum_{m=1}^{\infty} Y_m^{{''}}\frac{{m\pi}}{{a}}\cos \frac{{m\pi x}}{{a}}]^2 - 2(1-\nu)[\sum_{m=1}^{\infty} -Y_m(\frac{{m\pi}}{{a}})^2 \sin\frac{{m\pi x}}{{a}}][\sum_{p=1}^{\infty} Y_p^{{''}} \sin\frac{{p\pi x}}{{a}}]\} \ dxdy
- \int_{{0}}^{{b}} \int_{{0}}^{{a}} q \sum_{m=1}^{\infty} Y_m \sin \frac{{m\pi x}}{{a}} \ dxdy $$
    
Above equation reduces to,
$$ \pi = \frac{{D}}{{2}}\frac{{a}}{{2}} \int_{{0}}^{{b}} \{ \sum_{m=1}^{\infty} [-Y_m (\frac{{m\pi}}{{a}})^2 + Y_m^{{''}}]^2 + 2(1-\nu)[\sum_{m=1}^{\infty} (Y_m^{{'}})^2 (\frac{{m\pi}}{{a}})^2] - 2(1-\nu)[\sum_{m=1}^{\infty} (-Y_m)Y_m^{{''}} (\frac{{m\pi}}{{a}})^2] - \sum_{m=1}^{\infty} 2q_m Y_m \} \ dy $$
    
Here, $ q_m $ is defined as,
    
$$ q_m = \frac{{2}}{{aD}} \int_{0}^{a} q \sin \frac{{m\pi x}}{{a}} \ dx $$ <br>
By applying Euler-Lagrange equations for extremizing the above functional in the integration, <br>
$$ -\frac{{d^2}}{{dy^2}}(\frac{{\partial F}}{{\partial y_m^{{''}}}}) + \frac{d}{dy}(\frac{\partial F}{\partial Y_m^{'}})- \frac{\partial F}{\partial Y_m} = 0 $$
    
Where,
$$ F = \sum_{m=1}^{\infty} [-Y_m (\frac{{m\pi}}{{a}})^2 + Y_m^{{''}}]^2 + 2(1-\nu)[\sum_{m=1}^{\infty} (Y_m^{{'}})^2 (\frac{{m\pi}}{{a}})^2] - 2(1-\nu)[\sum_{m=1}^{\infty} (-Y_m)Y_m^{{''}} (\frac{{m\pi}}{{a}})^2] - \sum_{m=1}^{\infty} 2q_m Y_m $$
    
On solving the Euler-Lagrange equation, we get the following equation and boundary conditions
$$ Y_m^{''''} - 2 \left(\frac{m\pi}{a} \right)^2 Y_m^{''} + \left(\frac{m\pi}{a}\right)^4 Y_m = q_m $$
    
$$ Y_m^{''} - \nu \left(\frac{m \pi}{a} \right)^2Y_m = 0\ \ \ \ \ \ \ \ \ \ \ \ \ OR\ \ \ \ Y_m^{'} \ \ \ \ prescribed $$
    
$$ Y_m^{'''} - (2-\nu) \left(\frac{m \pi}{a} \right)^2Y_m^{'} = 0\ \ \ \ OR\ \ \ \ Y_m \ \ \ \ prescribed $$

# <font color="green"> Calculations

```python
import sympy as sp
from IPython.display import Math
```

```python
q0,a,b,m,D,h=sp.symbols('q0,a,b,m,D,h')
x=sp.symbols('x')
```

We'll have domain specific load equations as follows:
$$ q = \frac{2q_0x}{a}\ \ \ \ 0 \leq x\leq \frac{a}{2} $$
$$\hspace{1.5cm} =2q_0 - \frac{2q_0x}{a} \ \ \ \ \frac{a}{2} \leq x \leq a 
$$

Calculating $q_m$ using the formula,
$$ q_m = \frac{{2}}{{aD}} \int_{0}^{a} q \sin \frac{{m\pi x}}{{a}} \ dx $$ <br>

```python
2/(a*D)*(sp.integrate(2*q0/a*x*sp.sin(m*sp.pi*x/a),(x,0,a/2))+sp.integrate((-2*q0*x/a+2*q0)*sp.sin(m*sp.pi*x/a),(x,a/2,a))).simplify()
```

In above expression, $\sin \pi m $ will be zero for all integer values of m. It reduces to following expression

$$ q_m = \frac{{8q_0}}{{Dm^2\pi^2}}\sin \frac{{m\pi}}{{2}} $$

```python
qm=sp.symbols('qm')
```

```python
qm=8*q0*sp.sin(m*sp.pi/2)/(D*m**2*(sp.pi)**2)
```

```python
display(Math(r'q_m={}'.format(sp.latex(qm))))
```

```python
y=sp.symbols('y')
Y=sp.Function('Y')(y)
```

Now, solving the differential equation,
$$ Y_m^{''''} - 2 \left(\frac{m\pi}{a} \right)^2 Y_m^{''} + \left(\frac{m\pi}{a}\right)^4 Y_m = q_m $$
Substituting the expression of $q_m$ in above ODE, we get:

$$ Y_m^{''''} - 2 \left(\frac{m\pi}{a} \right)^2 Y_m^{''} + \left(\frac{m\pi}{a}\right)^4 Y_m = \frac{{8q_0}}{{Dm^2\pi^2}}\sin \frac{{m\pi}}{{2}} $$

```python
sol_temp=sp.dsolve(sp.diff(Y,y,4)-2*(m*sp.pi/a)**2*sp.diff(Y,y,2)+(m*sp.pi/a)**4*Y-qm)
```

After solving a differential equation we arrive at general solution of $Y_m(y)$:<br>
<b>$$ Y_m(y) = (C_{1}+C_{2}y)e^{\frac{-\pi my}{a}} + (C_{3}+C_{4}y)e^{\frac{\pi my}{a}} + \frac{8a^{4}q_{0}\sin(\frac{\pi m}{2})}{\pi^{6}Dm^{6}} $$</b>

Now applying boundary conditions to determine constants $C_{1}, C_{2}, C_{3}$ and $C_{4}$<br>

# <font color = "blue">Solving for constants

```python
C1,C2,C3,C4 = sp.symbols('C_1,C_2,C_3,C_4')
```

```python
sol =(C1+C2*y)*sp.exp(-sp.pi*m*y/a)+(C3+C4*y)*sp.exp(sp.pi*y*m/a)+8*a**4*q0*sp.sin(sp.pi*m/2)/(sp.pi**6*D*m**6)
```

# <font color = 'blue'>Boundary Conditions</font><br>
<h4>$1)$ $Y_m$ = 0 at y = 0</h4><br>
    $C_1 + C_3 + \frac{8a^{4}q_{0}\sin(\frac{\pi m}{2})}{\pi^{6}Dm^{6}} = 0$   <font color="brown">$- - - - - - - - (1)$ </font><br>

<h4>$2)$ $Y_m$ = 0 at y = b</h4><br> 
$(C_{1} + C_{2}b)e^{\frac{-\pi bm}{a}} + (C_{3} + C_{4}b)e^{\frac{\pi bm}{a}} + \frac{8a^{4}q_{0}\sin(\frac{\pi m}{2})}{\pi^{6}Dm^{6}} = 0$   <font color="brown">$- - - - - - - - (2)$ </font><br>

<h4>$3)\frac{dY_m}{dy}$ = 0 at y = 0 </h4><br> 
$\frac{-\pi C_{1}m}{a} + C_{2} + \frac{\pi C_{3}m}{a} + C_{4} = 0$   <font color="brown">$- - - - - - - - (3)$ </font><br>

<h4>$4)\frac{dY_m}{dy}$ = 0 at y = b </h4><br> 
$C_{2}e^{\frac{-\pi bm}{a}} + C_{4}e^{\frac{\pi bm}{a}} - \frac{\pi m(C_{1}+C_{2}b)e^{\frac{-\pi bm}{a}}}{a} + \frac{\pi m(C_{3}+C_{4}b)e^{\frac{\pi bm}{a}}}{a} = 0 $   <font color="brown">$- - - - - - - - (4)$ </font><br>

```python
eq1 = sp.Eq(sol.subs(y,0),0)
eq1
```

```python
eq2 = sp.Eq(sol.subs(y,b),0)
eq2
```

```python
eq3 = sp.Eq(sp.diff(sol,y).subs(y,0),0)
eq3
```

```python
eq4 = sp.Eq(sp.diff(sol,y).subs(y, b),0)
eq4
```

```python
result = sp.solve((eq1,eq2,eq3,eq4),(C1,C2,C3,C4))
```

After solving equations $(1),(2),(3),(4)$ we arrive at constants $C_{1},C_{2},C_{3},C_{4}$ in terms of available constants $q_{0},a,b$ and $m$ which is displayed further

```python
display(Math(r'C_1 = {}'.format(sp.latex(result[C1]))))
display(Math(r'C_2 = {}'.format(sp.latex(result[C2]))))
display(Math(r'C_3 = {}'.format(sp.latex(result[C3]))))
display(Math(r'C_4 = {}'.format(sp.latex(result[C4]))))
```

# <font color="blue">Corner Forces

Reaction forces at the edges is given by,<br>
<center>$ F_{x0} = \left( V_x - \frac{\partial M_{xy}}{\partial y}\right) $ at (0,y) </center><br>
<center>$ F_{xa} = \left( V_x - \frac{\partial M_{xy}}{\partial y}\right) $ at (a,y)</center><br>
<center>$ F_{y0} = \left( V_y - \frac{\partial M_{xy}}{\partial x}\right) $ at (x,0)</center><br>
<center>$ F_{yb} = \left( V_y - \frac{\partial M_{xy}}{\partial x}\right) $ at (x,b)</center><br>

$$ V_x = D\frac{\partial}{\partial x}\left(\frac{\partial^2 w}{\partial x^2} + \frac{\partial^2 w}{\partial y^2}\right) $$<br>
$$ V_y = D\frac{\partial}{\partial y}\left(\frac{\partial^2 w}{\partial x^2} + \frac{\partial^2 w}{\partial y^2}\right) $$<br>
$$ M_{xy} = -D(1-\nu)\frac{\partial^2 w}{\partial x\partial y} $$

```python
nu=sp.symbols('nu')
Fx0=(D*sp.diff(sp.diff(w,x,2)+sp.diff(w,y,2),x)+D*(1-nu)*sp.diff(sp.diff(w,x,y),y)).subs(x,0).subs(a,2).subs(nu,0.3).subs(D,1587.94).subs(m,1).evalf()
Fxa=(D*sp.diff(sp.diff(w,x,2)+sp.diff(w,y,2),x)+D*(1-nu)*sp.diff(sp.diff(w,x,y),y)).subs(x,2).subs(a,2).subs(nu,0.3).subs(D,1587.94).subs(m,1).evalf()
Fy0=(D*sp.diff(sp.diff(w,x,2)+sp.diff(w,y,2),y)+D*(1-nu)*sp.diff(sp.diff(w,x,y),x)).subs(y,0).subs(a,2).subs(nu,0.3).subs(D,1587.94).subs(m,1).evalf()
Fyb=(D*sp.diff(sp.diff(w,x,2)+sp.diff(w,y,2),y)+D*(1-nu)*sp.diff(sp.diff(w,x,y),x)).subs(y,2).subs(a,2).subs(nu,0.3).subs(D,1587.94).subs(m,1).evalf()
```

Integrating the force acting along $x=0$ line

```python
Rx0 = sp.integrate(Fx0,(y,0,2)).evalf()
Rx0
```

Integrating the force acting along $x=a$ line

```python
Rx2 = sp.integrate(Fxa,(y,2,0)).evalf()
Rx2
```

Integrating the force acting along $y=0$ line

```python
Ry0 = sp.integrate(Fy0,(x,0,2)).evalf()
Ry0
```

Integrating the force acting along $y=b$ line

```python
Ry2 = sp.integrate(Fyb,(x,2,0)).evalf()
Ry2
```

Net downward force acting

```python
F_down = (sp.integrate(sp.integrate(2*q0*x/a,(x,0,1))+sp.integrate(2*q0-2*q0*x/a,(x,1,2)),(y,0,2))).subs(q0,100).subs(a,2)
F_down
```

Net reaction force acting in upward direction

```python
F_up = Rx0 + Rx2 + Ry0 + Ry2
F_up
```

Net reaction forces generated at corners in upward direction are

```python
F_net = F_up + F_down
F_net
```
